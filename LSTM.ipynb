```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas_datareader import data
```

```python
nse = pd.read_excel('Final Dataset1.xlsx', sheet_name='NIFTY')
```

```python
nse.drop(nse.columns[[6]], axis=1, inplace=True)
nse.head()
```

```python
nse = nse.rename(columns={'Shares Traded': 'Volume'}, index={'ONE': 'Row_1'})
nse.head()
```

```python
nse.describe()
```

```python
nse1 = nse.loc[nse['Date']<='2020-03-31']
```

```python
nse1.set_index(pd.DatetimeIndex(nse1['Date']), inplace=True)
nse1
```

```python
nse1.drop(nse1.columns[[0]], axis=1, inplace=True)
nse1.head()
```

```python
stock_data_len = nse1['Close'].count()
print(stock_data_len)
```

```python
training_set = nse1.iloc[:,3].values
```

```python
training_set = np.array(training_set)
training_set = np.reshape(training_set, (733,1))
```

```python
print(training_set)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)
print(training_set_scaled.shape)
```

```python
features = []
labels = []
for i in range(60, stock_data_len):
    features.append(training_set_scaled[i-60:i, 0])
    labels.append(training_set_scaled[i, 0])

features = np.array(features)
labels = np.array(labels)

features = np.reshape(features, (features.shape[0], features.shape[1], 1))
```

```python
print(labels)
```

```python
print(features)
```

```python
print(features.shape)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(units = 50, return_sequences = True, input_shape = (features.shape[1], 1)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(units = 50, return_sequences = True),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(units = 50, return_sequences = True),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(units = 50),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(units = 1)
])
```

```python
print(model.summary())
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
from time import time
start = time()
history = model.fit(features, labels, epochs = 20, batch_size = 32, verbose = 1)
end = time()
```
